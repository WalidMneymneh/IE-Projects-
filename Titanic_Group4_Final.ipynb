**Question 3:**
Compute the percentage of survivors by sex (i.e. the percentage of male passengers that survived and female passengers that survived) and the sex distribution of survivors (i.e. percentage of survivors that were men and percentage of survivors that were women). Paste the answers below.


```python
# Simplifying task with function that takes two values: 
# sex refers to what sex's survival rate you want
# population refers to out of what people you want to do that calculation, i.e. male survivors out of all male passengers

def survival_rates(sex,population):
    if population == "survivors":
        return round(100* len(df.loc[(df["Survived"] == 1 )&(df["Sex"] == sex )]) / len(df.loc[(df["Survived"] == 1 )]),2)
    else:
        return round(100* len(df.loc[(df["Survived"] == 1 )&(df["Sex"] == sex )]) / len(df.loc[(df["Sex"] == population )]),2)  
```

```python
# survival rate of males
survival_rates("male","male")
```

```python
# To cross check whether the formula works for male survivors out of all males: 
round(100* len(df.loc[(df["Survived"] == 1 )&(df["Sex"] == "male" )]) / len(df.loc[(df["Sex"] == "male" )]),2)
```

```python
# survival rate of females
survival_rates("female","female")
```

```python
# percentage of male survivors out of all survivors
survival_rates("male","survivors")
```

```python
# To cross check whether the formula works for all male survivors out of all survivors 
round(100* len(df.loc[(df["Survived"] == 1 )&(df["Sex"] == "male" )]) / len(df.loc[(df["Survived"] == 1 )]),2)
```

```python
# percentage of female survivors out of all survivors
survival_rates("female","survivors")
```

```python
# Brining the survivors table into right format
survival_gender.loc[["male","female"],["Survived","Not Survived"]]
```

(2 Points)

Devise two distinct strategies to impute the missing values in the Age column.
 
(You cannot use machine learning algorithms for the feature engineering part, only manual techniques. This rules out KNN, Deep Learning, sklearn.preprocessing.Imputer, fancyimpute, missingno, and similar.)
 
Some basic ideas (that must not be reused), in increasing order of sophistication:
 
    Fill all the missing values with random values (very bad!)
    Fill all the missing values with the average age of the whole dataset (bad!)
    Fill all the missing values with the median age (so so)
    Fill all the missing values for men with the median age of men, and the missing values of women with the median age of women (better)
 
The objective is to find groupings that yield more granular information about the age.
 
Extra points awarded for proper justifications of the choices made in terms of some quantitative metric (for example, the standard deviation of the age for each sub-group).

```python
# Getting overview of number of missing values
len(df.loc[df["Age"].isnull()])
```

```python
# Our starting point is gender as grouping variable. We verify, that we have no missing values in the gender column.
df["Sex"].unique()
```

In order to evaluate the suitability of the sub-group for predicting the age, we need to understand the baseline standard deviation for age based on just gender first. 
Any subgroup we create should ideally have a **smaller standard deviation** (~ <14) than if we just group by gender in order to be a better predictor.

```python
# Relevant descriptive statistics for gender to understand our baseline
df.pivot_table(
    index=["Sex"],
    values="Age",
    aggfunc=["mean","median","std"])
```

---
**Strategies & Approach**

Our hypothesis is that the following additional variables will help better predict the age of the passengers in addition to gender:


*   **Title**: The title can indicate whether someone is a child ("Master") or unmarried (usually younger than married) 
*   **Pclass**: The passenger class on the ship is highly correlated with the fare price and serves as a categorical proxy for the societal level that a passenger is in. It may help predict age as there may be differences in the median age between different societal levels
*   **Survived**: Whether a person survived or not may give additional insight into how old a person was (although only in combination with gender, Title and Pclass)



We will use these variables to iteratively improve our prediction power for the missing age values.

---
**Creating first variable "Title" and checking relevance:**

```python
# First, we verify, that the names do not contain null values, because we will be extracting the titles from the names
df["Name"].isnull().value_counts()
```

```python
# Create the new variable using the first part of the name before the point (.)
df["Title"] = df["Name"].str.split(".").str[0]
df.head()
```

```python
# We make sure that the titles have been correctly extracted and confirm, that all entries in column "Title" are indeed titles
df["Title"].value_counts()
```

**Example 1: Masters**

We verify, that for all "Masters", the standard deviation of age is indeed much smaller than for just gender. Note that Masters can only be male, so we did not need to filter for gender here as it would give us the same result.

```python
# Standard deviation for age for people with title "Master"
df.loc[df["Title"]=="Master"]["Age"].describe()["std"]
```

---
**Confirming first variable "Title":**
* By adding the "Title" variable to the grouping condition, we can effectively reduce the standard deviation of age for all subgroups
* Because standard deviation cannot be computed if a group only has one value, we don't see all titles in this table
* We will check whether the missing values apply to other titles than these later on

```python
# Standard deviation for groups by gender and title to compare with baseline stdv of around 14 for just gender
df.pivot_table(
    index=["Sex","Title"],
    values="Age",
    aggfunc=["std"])
```

---
**Confirming second variable "Pclass":**
* At this point, we add another variable to the grouping conditions, namely "Pclass"
* We hypothesize that there is an age difference between the different ticket classes which are a proxy for societal classes
* Indeed, we see that for some sub-groups of title and class, the standard deviation of age can be further reduced

```python
# Standard deviation for groups by gender, title and pclass to compare with standard deviation with just gender & title
df.pivot_table(
    index=["Sex","Title"],
    columns=["Pclass"],
    values="Age",
    aggfunc=["std"])
```

**Cross-check with "fare" - could it be useful to include fare as well?**
* It might be interesting to additionally use fare as a predictor for societal level and thus improve our predictive power for age
* But looking at the fares within the different Pclasses, we see a (non-statistically speaking) significant difference between the fares
* Since we would need to create a categorical variable for fare in order to use it for grouping, we decide to use Pclass as a reasonable proxy instead
* With Pclass, we can most likely capture most of the relevant correlation between societal level and age that could be derived from "fare"

```python
# Checking correlation between pclass and fare
df.pivot_table(
    index=["Pclass"],
    values="Fare",
    aggfunc=["max","min","mean","median","std"])
```

---
***Not* onfirming third variable "Survived":**
* Finally, we add "Survived" as a differentiator to our existing groups to check whether it improves the prediction power for age
* Unfortunately, the variable does not provide additional insight into the age gap as it either results in similar or higher standard deviation for most groups
* This may be due to the fact that there is no clear relationship between surviving and gender, title or pclass in that combination
* Including "Survived" would therefore in this variable combination diminish our predictive power for age. We chose to not include it in the final set


```python
# Checking standard deviation for all sub-groups gender, title, pclass and survived to see whether "survived" improves standard deviation
df.pivot_table(
    index=["Sex","Title"],
    columns=["Pclass","Survived"],
    values="Age",
    aggfunc=["std"])
```

---
**Final Sub-Group Variables: Gender, Title and Pclass (Part 1): Mean and Median Comparison**
* Here, we compute the mean and median for the sub-groups created from the final "prediction" variables: gender, title and pclass
* We see that the mean and median are in most cases very close together or even exactly the same
* It is most important to check this for the sub-groups that still posess a rather larger standard deviation for age, namely for any "Mrs", "Mr" and "Rev"
* Although there is only a small difference between mean and median, we choose to use the median to impute age after all

```python
# Age mean and median by gender, title and pclass to check difference between mean and median
df.pivot_table( 
    index=["Sex","Title"],
    columns=["Pclass"],
    values="Age",
    aggfunc=["mean","median"])
```

---
**Final Sub-Group Variables: Gender, Title and Pclass (Part 2): Checking Sub-Groups for Missing Age Values**

```python
# Finally, we check which sub-groups are actually missing age values.
df.loc[df["Age"].isnull()].pivot_table(
    index=["Sex","Title"],
    columns=["Pclass"],
    values="Name",
    aggfunc=["count"])
```

```python
# Checking number of records for relevant missing gender, title and pclass sub-groups (Miss, Mrs, Dr, Master, Mr)
prediction_data_count = df.pivot_table(
    index=["Sex","Title"],
    columns=["Pclass"],
    values="Age",
    aggfunc=["count"])
prediction_data_count.loc[[("female","Miss"),("female","Mrs"),("male","Dr"),("male","Master"),("male","Mr")]]
```

* We verify that we have enough data to predict the missing ages for the given sub-groups
* Indeed, for all sub-groups, we have existing passenger data to infer an age for the missing values (e.g. 45 people with title "Miss" and Pclass of "1")

---
**Final Sub-Group Variables: Gender, Title and Pclass (Part 3): Creating Final Age Imputation Table**

* Finally, we create the imputation table which we would use to impute the missing values
* We see that this creates a significantly different picture than if we had just used the median ages of 27 (female) and 30 (male) to impute the missing values
* For example, young masters will be much younger and there is a difference in age for the "Missuses" and "Mrs" depending on the Pclass they belong in
* In general, adults and children seem to have significantly different ages - we shall test this theory in the next part

```python
# Median table for relevant missing gender, title and pclass sub-groups (Miss, Mrs, Dr, Master, Mr)
imputation_table = df.pivot_table(
    index=["Sex","Title"],
    columns=["Pclass"],
    values="Age",
    aggfunc=["median"])
imputation_table.loc[[("female","Miss"),("female","Mrs"),("male","Dr"),("male","Master"),("male","Mr")]]
```

---
**Additional Cross-Check: Adult vs. Child Theory**

```python
# Creating another variable for being an adult or a child 
df["Is_Adult"] = df["Age"]>18
df.head()
```

```python
# median and standard deviation for gender and "is_adult" variables to check whether "is_adult" decreases standard deviation
df.pivot_table(
    index=["Sex"],
    columns=["Is_Adult"],
    values="Age",
    aggfunc=["median","std"])
```

* We verify that indeed, the median age differs heavily by this categorization
* The standard deviation is also much lower (than initial 14) within both the "kids" and "adults" group, confirming that it is important to make the distinction between kids and adults when imputing age values

```python
# Combining "is_adult" variable with our other variables to confirm that standard deviation is improved when introducing the "is_adult" variable
df.pivot_table(
    index=["Sex","Title"],
    columns=["Is_Adult","Pclass"],
    values="Age",
    aggfunc=["std"])
```

* We further verify that this also holds true when we add the "Adult" variable to our sub-groups
* We see that the title is a good indicator for the age, given that masters can only be kids, "Cols", "Drs", "Majors" and "Revs" can only be adults
* Pclass further helps making the imputation more precise where we have no clear distinction between adults and kids just from the title
